```python
import yfinance as yf
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
tickers = [
    # equities
    "VOO",
    "ISF.L",
    "2800.HK",
    "VXX",
    "VXZ",
    # commodities
    "CL=F",
    "GC=F",
    "HG=F",
    "NG=F",
    # interest rates
    "^IRX",
    "^FVX",
    "^TNX",
    "^TYX",
    # fx
    "EURUSD=X",
    "SGD=X",
    "CNY=X",
    "JPY=X",
    # crypto
    "BTC-USD",
]
```

```python
start_date  = "2024-03-14"
end_date    = "2024-06-14"
```

```python
df = yf.download( tickers=tickers, start=start_date, end=end_date)
df.head()
```

```python
col = 'Adj Close'
sdf = ( df[col] - df[col].mean() ) / df[col].std()
sdf.fillna(method='ffill', inplace=True)
sdf.head()
```

```python
sdf[['2800.HK', 'JPY=X', 'VOO']].plot()
```

```python
sdf.corr()
```

```python
1 - sdf.corr()
```

```python
sdf[['BTC-USD', 'SGD=X', '2800.HK']].plot()
```

```python
rdf = pd.DataFrame()
for ticker in tickers:
    rdf[ticker] = sdf[ticker] - sdf[ticker].shift(1)
    
rdf.dropna(axis=0, inplace=True)
rdf.head(20)
```

```python
rdf.corr().sort_index(axis=0).sort_index(axis=1)
```

```python
sdf.corr().sort_index(axis=0).sort_index(axis=1)
```

```python
sdf[['VOO', 'SGD=X']].plot()
```

```python
date = '2024-06-13'
directions = rdf[rdf.index == date].to_dict(orient='records')[0]
rdf.loc[date].sort_values().plot(kind='barh')
display(dict(sorted(directions.items(), key=lambda x: x[1])))
```
